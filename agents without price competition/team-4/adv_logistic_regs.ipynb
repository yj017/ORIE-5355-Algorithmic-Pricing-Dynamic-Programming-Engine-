```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import pickle
import os
```

```python
DATA_PATH = '../../data/train_prices_decisions_2025.csv'
df = pd.read_csv(DATA_PATH)
y = df['item_bought'] 

df['price_x_cov2'] = df['price_item'] * df['Covariate2']
df['price_x_cov1'] = df['price_item'] * df['Covariate1']
df['price_squared'] = df['price_item']**2
df['log_Covariate1'] = np.log1p(df['Covariate1'])
df['log_Covariate2'] = np.log1p(df['Covariate2'])


# B
features_B = ['Covariate1', 'Covariate2', 'Covariate3', 'price_item', 'price_x_cov2']
X_B = df[features_B]

model_B = LogisticRegression(max_iter=1000)
model_B.fit(X_B, y)

with open('adv_logB_model.pkl', 'wb') as f:
    pickle.dump(model_B, f)

# C
features_C = ['Covariate1', 'Covariate2', 'Covariate3', 'price_item', 
              'price_x_cov2', 'price_x_cov1']
X_C = df[features_C]

model_C = LogisticRegression(max_iter=1000)
model_C.fit(X_C, y)

with open('adv_logC_model.pkl', 'wb') as f:
    pickle.dump(model_C, f)

# D
print("\nTraining Model D (Complex)...")
features_D = ['Covariate1', 'Covariate2', 'Covariate3', 'price_item', 
              'price_x_cov2', 'price_squared', 'log_Covariate1', 'log_Covariate2']
X_D = df[features_D]

model_D = Pipeline([
    ('scaler', StandardScaler()),
    ('logit', LogisticRegression(max_iter=1000))
])
model_D.fit(X_D, y)

with open('adv_logD_model.pkl', 'wb') as f:
    pickle.dump(model_D, f)
```
