```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb 
import pickle
import os
```

```python
DATA_PATH = '../../data/train_prices_decisions_2025.csv'
RF_MODEL_PATH = 'rf_model.pkl'
XGB_MODEL_PATH = 'xgb_model.pkl'

df = pd.read_csv(DATA_PATH)

features = ['Covariate1', 'Covariate2', 'Covariate3', 'price_item']
X = df[features]
y = df['item_bought']

# rf
model_E = RandomForestClassifier(n_estimators=100, max_depth=10, n_jobs=-1, random_state=42)
model_E.fit(X, y)

with open(RF_MODEL_PATH, 'wb') as f:
    pickle.dump(model_E, f)

# xgb
model_F = xgb.XGBClassifier(n_estimators=100, max_depth=5, learning_rate=0.1, 
                            n_jobs=-1, random_state=42, 
                            use_label_encoder=False, eval_metric='logloss')
model_F.fit(X, y)

with open(XGB_MODEL_PATH, 'wb') as f:
    pickle.dump(model_F, f)
```
